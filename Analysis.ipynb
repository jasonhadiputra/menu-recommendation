## Insights and Potential Features

### User centric

Since the analysis is user-based (i.e., given `user_id` and past sales behavior), we will need to feature engineer on user level. To use content-based filtering, we will need to create a metric to measure the liking of a user to a menu.

The metric I choose is count of menu. The idea is that if someone doesn't like what they order, they wouldn't order again. So the more menu they order, the more they like the menu.

However, if the user only ordered once, it is hard to tell whether the user likes the menu or not. If it has been a long time since they ordered the menu, we can say that they don't like the menu they only ordered once.

Therefore, we need to engineer two features
1. `avg_trx_interval`
2. `rating`

On top of that, note that some menus are similar. Therefore, we can engineer new grouping for the menus.

#### 1. `avg_trx_interval`

How long it has been since the user ordered the menu? I choose to use the average of the time interval between orders.

The idea is that if the average time interval is long, it means that the user doesn't order frequently.

If the average time interval is short, it means that the user order frequently.

```python
df_trx_interval = (
    df_sales
    .merge(
        df_bridge[['sales_id','user_id']].drop_duplicates(),
        on='sales_id',
        how='left',
        )
    .sort_values(['user_id','trx_date_detail'])
    .assign(
        # Calculate the interval between transactions
        avg_trx_interval=lambda x: x.groupby('user_id')['trx_date_detail'].diff(),
        )
    .groupby('user_id')
    .agg({'avg_trx_interval': ['mean']})
    .droplevel(1, axis=1)
    .reset_index()
    )

# For idempotency
if 'avg_trx_interval' in df_user.columns:
    df_user = df_user.drop(columns='avg_trx_interval')

df_user = df_user.merge(df_trx_interval, on='user_id')
df_user
```

Some people haven't ordered more than once and as such have no `avg_interval`. I decide to impute them with the corresponding mean `avg_interval` of their `user_tier_level` and `user_gender` since there is a quite significant difference between them.

```python
(df_user
 .groupby(['user_tier_level', 'user_gender'])
 .agg({'avg_trx_interval': ['mean']})
#  .reset_index()
#  .droplevel(1, axis=1)
 )
```

```python
df_user = (df_user
           .merge((df_user
                   .groupby(['user_tier_level', 'user_gender'])
                   .agg({'avg_trx_interval': ['mean']})
                   .reset_index()
                   .droplevel(1, axis=1)),
                  on=['user_tier_level', 'user_gender'],
                  suffixes=['','_mean'],
                  )
           .assign(avg_trx_interval=lambda x: (
                   x['avg_trx_interval']
                   .fillna(x['avg_trx_interval_mean'])
                   ))
           .drop('avg_trx_interval_mean', axis=1)
           )

df_user
```

#### 2. `df_rating`

How many times the user ordered the menu? I define `rating` := **sum of `quantity` of each `menu_id`**.

The idea is that if the user ordered a lot, it means that the user likes the menu.

If it has been a long time since the user ordered the menu, it means that the user doesn't like the menu and we will not recommend the menu to the user.

We will also assign the negative of the rating to indicate that the user doesn't like the menu. A user is said not to like a menu if it has been more than their average interval time since the last time they ordered the menu.

```python
df_rating = (
      df_bridge
      .merge(df_user, on='user_id')
      .merge(df_sales, on='sales_id')
      .groupby(['user_id', 'menu_id'])
      .agg({'quantity': 'sum',
            # Time since last order
            'trx_date_detail': lambda x: NOW - x.max()})
      .reset_index()
      .merge(df_user, on='user_id')
      .assign(
            rating=(
                  # Do for each user
                  lambda x: x.apply(
                        lambda y: (
                              y['quantity']
                              # Time since last purchase is short
                              if y['trx_date_detail']<y['avg_trx_interval']
                              # or quantity is more than 1
                                    or y['quantity']>1
                              else -y['quantity']
                              ),
                        axis=1
                        )
                  )
            )
      .loc[:, ['user_id', 'menu_id', 'rating']]
      )

df_rating
```

# Modeling

My idea is to use content-based filtering. As such, we will need to create two dataframes. One for user and one for menu.

```python
user_cols = [
    'user_id',
    'user_tier_level',
    'user_gender',
    'time_since_created',
    'avg_trx_interval',
    ]

df_user_modeling = (df_user
 .loc[:, user_cols]
 .set_index('user_id')
 .assign(
     user_gender_M=lambda x: x['user_gender'].map({'M': 1, 'F': 0}),
     user_tier_level=lambda x: x['user_tier_level'].cat.codes,
     time_since_created=lambda x: x['time_since_created'] / pd.to_timedelta(1, unit='W'),
     avg_trx_interval=lambda x: x['avg_trx_interval'] / pd.to_timedelta(1, unit='W'),
     )
 .drop('user_gender', axis=1)
 )

df_user_modeling
```

```python
menu_cols = [
    'menu_id',
    'brand',
    'menu_category',
    'menu_category_detail',
    # 'menu_name',
    'last_sold',
    'avg_sales_interval',
]

df_menu_modeling = (pd.get_dummies(
    df_menu
    .loc[:, menu_cols]
    .set_index('menu_id')
    )
    .assign(
        last_sold=lambda x: x['last_sold'] / pd.to_timedelta(1, unit='W'),
        avg_sales_interval=lambda x: x['avg_sales_interval'] / pd.to_timedelta(1, unit='W'),
        )
 )

df_menu_modeling
```

```python
from sklearn.model_selection import train_test_split

input_user = (
    df_rating
    .set_index('user_id')
    .join(df_user_modeling)
    .drop(columns=['menu_id', 'rating'])
    .to_numpy()
    )

input_menu = (
    df_rating
    .set_index('menu_id')
    .join(df_menu_modeling)
    .drop(columns=['user_id', 'rating'])
    .to_numpy()
    )

y = df_rating['rating'].to_numpy()

(user_train, user_test,
 menu_train, menu_test,
 y_train, y_test) = train_test_split(input_user,
                                     input_menu,
                                     y,
                                     test_size=0.2,
                                     random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras

num_outputs = 32
tf.random.set_seed(42)

num_user_features = input_user.shape[1]
num_menu_features = input_menu.shape[1]

user_NN = tf.keras.models.Sequential([     
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(num_outputs),  
])

menu_NN = tf.keras.models.Sequential([     
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(num_outputs),  
])

# create the user input and point to the base network
input_user = tf.keras.layers.Input(shape=(num_user_features))
vu = user_NN(input_user)
# vu = tf.linalg.l2_normalize(vu, axis=1)

# create the menu input and point to the base network
input_menu = tf.keras.layers.Input(shape=(num_menu_features))
vm = menu_NN(input_menu)
# vm = tf.linalg.l2_normalize(vm, axis=1)

# compute the dot product of the two vectors vu and vm
output = tf.keras.layers.Dot(axes=1)([vu, vm])

# specify the inputs and output of the model
model = tf.keras.Model([input_user, input_menu], output)

model.summary()
```

```python
cost_fn = tf.keras.losses.Huber()
model.compile(optimizer='adam',
              loss=cost_fn)

model.fit([user_train, menu_train], y_train, epochs=30)
```

There's a slight overfit but I will keep it. Will be updated if I find a better model.
